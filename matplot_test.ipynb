```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
meses = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho']
valores = [8954, 9503, 11947, 10494, 11483, 12949]
```

image.png


Plotagem dos gráficos:

```python
plt.bar(meses,valores) # Define os eixos do gráfico do tipo 'Barras'

menor = valores < np.average(valores)
maior = valores > np.average(valores)

media = np.average(valores)
plt.axhline(y=media, color='r', linestyle='-') # Plotando linha de média de faturamento
plt.title('Acumulado em Vendas por Mês')
plt.xlabel('Meses') 
plt.ylabel('Valores')
plt.show()

```

Criando segunda visualização dos dados no modelo Pie Chart.

```python
def perc(pct, values): # Retorna o percentual de faturamento do mês
  inteiro = int(pct/100 * sum(values))
  return "{:.1f}%".format(pct)

plt.pie(valores, labels=meses, textprops=dict(color='w'),autopct=lambda pct: perc(pct, valores))
plt.title('Faturamento por Mês')
plt.legend(meses, title='Meses', loc='center right', bbox_to_anchor=(1, 0, 0.5,1))
plt.show()
```

```python
media = np.average(valores)

def perc(pct, values): # Retorna o percentual de faturamento do mês
  inteiro = int(pct/100 * sum(values))
  return "{:.1f}%".format(pct)

plt.subplot(1,2,1) # Primeiro Plot
plt.bar(meses,valores) # Define os eixos do gráfico do tipo 'Barras'
plt.axhline(y=media, color='r', linestyle='-') # Plotando linha de média de faturamento
plt.title('Acumulado em Vendas por Mês')
plt.xlabel('Meses') 
plt.xticks(rotation=90) #Rotaciona os rótulos do Eixo X em 90°
plt.ylabel('Valores')

plt.subplot(1,2,2) # Segundo plot
plt.pie(valores, labels=meses, textprops=dict(color='w'),autopct=lambda pct: perc(pct, valores))
plt.title('Faturamento por Mês')
plt.legend(meses, title='Meses', loc='center right', bbox_to_anchor=(1, 0, 0.5,1))
plt.show()

```
